Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a maximum score of 800 points. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()
```

```python
# Question 1 - create DataFrame identifying schools where average math SAT score is at least 80% of the maximum possible score (800)
top_scores = 0.80*800
best_math_schools = schools[schools["average_math"] >= top_scores]
best_math_schools = best_math_schools[["school_name","average_math"]]
best_math_schools = best_math_schools.sort_values(by=['average_math'], ascending=False)
best_math_schools
```

```python
# Question 2 - identify top 10 performing schools based on scores across the three SAT sections
schools["total_SAT"] = schools["average_math"] + schools["average_reading"] + schools["average_writing"]
schools_sort = schools.sort_values(by=['total_SAT'],ascending=False)
top_10 = schools_sort.head(10)
top_10_schools = top_10[["school_name","total_SAT"]]
top_10_schools
```

```python
# Question 3 - locate NYC borough with largest standard deviation for total_SAT and store as a DataFrame with three columns: num_schools, average_SAT, std_SAT

# First, identify which borough has largest standard deviation: Manhattan
schools.groupby("borough").total_SAT.std().sort_values(ascending=False)
```

```python
# Isolate Manhattan and take a look at values of interest
Manhattan = schools[schools["borough"] == 'Manhattan']
Manhattan["total_SAT"].agg(['count','mean','std'])
```

```python
# Create the requested DataFrame
count = Manhattan.school_name.count()
std = Manhattan.total_SAT.std().round(2)
average = Manhattan.total_SAT.mean().round(2)

largest_std = [['Manhattan',count,average,std]]

largest_std_dev = pd.DataFrame(largest_std, columns = ['borough','num_schools','average_SAT','std_SAT'])

largest_std_dev.set_index('borough', inplace=True)

largest_std_dev
```
