```python
import pandas as pd 
lego_sets = pd.read_csv("datasets/lego_sets.csv")
parent_themes = pd.read_csv("datasets/parent_themes.csv")
```

```python
# Remove entries that have invalid set_num, since set_num is a critical field
mod_lego = lego_sets[lego_sets.set_num.notnull()]
```

```python
# Check to see if any Lego sets have null parent themes
mod_lego[mod_lego.parent_theme.isnull()]
```

```python
# The business questions are solely about licensed sets, so unlicensed parent themes can be filtered out
lic_parent = parent_themes[parent_themes.is_licensed == True]
```

```python
lic_parent.rename(columns={'name': 'parent_theme'}, inplace=True)
```

```python
# Again, we only care about licensed sets, so we do an inner join to remove data from mod_lego that is on unlicensed sets
merged = pd.merge(mod_lego, lic_parent, how='inner')
```

Question 1: What percentage of all licensed sets ever released were Star Wars themed?

```python
all_licensed = merged.set_num.count()
```

```python
Star_Wars = merged[merged['parent_theme'] == 'Star Wars'].set_num.count()
```

```python
the_force = (Star_Wars / all_licensed)*100
```

```python
# the_force is currently a float - we need it to be an integer
print(type(the_force))
# Cast as an integer
the_force = the_force.astype('int')
```

```python
# Check that the_force is now an integer
type(the_force)
```

Question 2: In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

```python
# Create a Series that depicts the maximum count of sets of an individual parent theme by year
num_by_year = merged.groupby(["year","parent_theme"]).set_num.count()
max_by_year = num_by_year.groupby("year").max()
```

```python
# Create a dictionary using year as key and number of sets as value
dict = {}
for i in range(len(max_by_year)):
    year = 1999 + i*1
    num = max_by_year.iloc[i]
    dict[year] = num
print(dict)
```

```python
# Repeat the above, but just for Star Wars themed sets
star_wars_count = merged[merged['parent_theme'] == 'Star Wars'].groupby(["year"]).set_num.count()
```

```python
sw_dict = {}
for i in range(len(star_wars_count)):
    year = 1999 + i*1
    num = star_wars_count.iloc[i]
    sw_dict[year] = num
print(sw_dict)
```

```python
# Find which year has a mismatch between Star Wars sets count and max sets count
for key,val in dict.items():
    for key2, val2 in sw_dict.items():
        if key==key2 and val!=val2:
            print(key)
```
